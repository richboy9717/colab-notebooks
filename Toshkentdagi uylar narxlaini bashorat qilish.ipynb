```python
from sklearn.model_selection import train_test_split

strat_train_set , strat_test_set = train_test_split(df, test_size=0.2, random_state=17)

X_train = strat_train_set.drop(['price'],axis=1) # Label (Bizni Modelimiz bashorot qilishi kerak bo'lgan ustun) ni tashlab yubordik
y = strat_train_set['price'].copy() # #Labelni alohida o'zgaruvchiga olib qo'ydik

X_num = X_train.drop(['address','district'], axis=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder, StandardScaler

num_pipeline = Pipeline([
          ('std_scaler', StandardScaler())             
])
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(X_num)
cat_attribs = ['address','district']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OrdinalEncoder(), cat_attribs)
])
```

```python
X_prepared = full_pipeline.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression

LR_model = LinearRegression()
```

```python
LR_model.fit(X_prepared, y)
```

```python
x_test = strat_test_set.drop(['price'], axis=1)
y_test = strat_test_set['price'].copy()

prepared = full_pipeline.fit_transform(x_test)

predicted = LR_model.predict(prepared)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predicted)
print('RMSE: ' , np.sqrt(mse))
mae = mean_absolute_error(y_test, predicted)
print('MAE: ' , mae)
```

#RandomForest

```python
from sklearn.ensemble import RandomForestRegressor

RF_model = RandomForestRegressor()

RF_model.fit(X_prepared,y)
```

```python
x_test = strat_test_set.drop(['price'], axis=1)
y_test = strat_test_set['price']

prepared = full_pipeline.fit_transform(x_test)

predicted = RF_model.predict(prepared)
```

```python
mse = mean_squared_error(y_test, predicted)
print('RMSE: ' , np.sqrt(mse))

mae = mean_absolute_error(y_test, predicted)
print('MAE: ' , mae)
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
Tree_model = DecisionTreeRegressor()
Tree_model.fit(X_prepared, y)
```

```python
x_test = strat_test_set.drop(['price'], axis=1)
y_test = strat_test_set['price']

prepared = full_pipeline.fit_transform(x_test)

predicted = Tree_model.predict(prepared)
```

```python
mse = mean_squared_error(y_test, predicted)
print('RMSE: ' , np.sqrt(mse))

mae = mean_absolute_error(y_test, predicted)
print('MAE: ' , mae)
```
